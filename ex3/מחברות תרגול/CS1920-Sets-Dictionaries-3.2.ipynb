#### Hashing
- A hash function is a function that can be used to map data of arbitrary size to fixed-size values. 
- The values can be used to index data of different sizes into a fixed-size table (hash table). 
- Use of a hash function to index a hash table is called hashing.

#### The *`hash(object)`* function (in Python)

• Returns the `hash value` of the object (if it has one).<br>
• Hash values are integers:<br>
• They are used to quickly compare dictionary keys during
a dictionary lookup. <br>
• Numbers, Strings, Tuples (and other objects..) are Hashable ( have hash values )<br>
• Numbers that compare equal have the same hash value<br>
  (even if they come from different classes: `1 (int)`, and `1.0 (float)`<br>

<h2><center>Operations with sets: the results are sets</center></h2>

### Set difference `s1-s2`, between two sets `s1`, `s2`  
#### the result is a set whose elements are in `s1` but not in `s2`

```python
s1 = set('abcd') # {'a', 'b', 'c', 'd'}
s2 = set('cdef') # {'c', 'd', 'e', 'f'}

# the sets difference operation
s1 - s2 # in s1 but not in s2 : {'a', 'b'}
```

### `s1 | s2` : union of sets `s1`,  `s2` 
**{'a', 'b', 'c', 'd', 'e', 'f'}**


```python
# sets union (איחוד בין שתי קבוצות)
s1 | s2
```

### s1 & s2 : Intersection of s1 ,s2

```python
s1 & s2 #{'c', 'd'}
```

### s1^s2 : symmetric difference:
all elements that are in `s1` or in `s2` but *not in both*

```python
# symmetric difference:
s1^s2 # in s1 or in s2 but not in both :{'a', 'b', 'e', 'f'}
```

### Like other iterable collections, sets support:
- membership testing : x `in` set, x `not in` set,
- finding the number of elements `len(set)`,
- for loops : `for x in set`: 
- Being an unordered collection, sets *do not record*: element position, nor order of insertion.



```python
s2 = set('Halva')
print(len(s2))

for i in s2:
    print(i,end =' ')
    
```

### Sets do not support:
- indexing, slicing, or
- other **sequence-like** behavior

### In other languages, this kind of data structure is called sometimes *associative memory*, or *associative array*

#### About Dictionary
• A dictionary is an *unordered associative container* that maps `keys` to `values`<br>
• Items in a dictionary always exist as `key : value` *pairs* <br>
• A key can be any *hashable (immutable)* object like: `integer`, `string`, `tuple`, `frozenset` etc.<br>
• A value can be any object :  `integer`, `string`, `tuple`, `list`, `set`, `frozenset`, `dictionary` etc. <br>
• A key is unique in a dictionary

| key     |   value    |
|---------|------------|
| 'a'   : | [1,2,3]   |
| 12    : | 'abc'      |
| (2,'b'):| {1,2,3}    |

### Creating dictionaries :

```python
# Create an empty dictionary!
A = {}
print('A =', A)

# Set initial key:value pairs
B = {'a' : 'alef', 'd' : 'dalet', 'g' : 'ghimel'}
print('B =', B)

C = {(1,2) : False, "bar" : [1, 7, 3], 2.99 : "almost three"}
print('C =', C)

```

### Update a dict with data from another dictionary
#### To merge two dictionaries we can use `update()`
Example:

```python
A = {1: 20, -5: 7, 8.2: 31}
print('A =', A)

B = {1:'foobar', 8.2:0}
print('B =', B)

A.update(B) # A is updated with pairs from B
print ('The New updated A =', A) # the pair 8.2:31, is updated to 8.2:0

print('There are', len(A), 'pairs in the New Updated A') # 4 key : values pairs
```

### To remove a pair and get both the key and the value : `popitem()`¶ <br>
Remove and return a `(key, value)` pair from the dictionary. <br>

From python3.7 Pairs are returned in **LIFO** order (*Last In First Out*).

`popitem()` is useful to destructively iterate over a dictionary. 

If the dictionary is `empty`, calling `popitem()` raises a `KeyError`.

Example:


```python
d = {1:'a', 'b': 'iosi', 3:[22,11], (1,):()}
print('The Original d =', d)

for i in range(len(d)):
    k, v = d.popitem() # does not use a Key 
    print('k =', k, ", v=", v)
    print('d =', d)
```

<h1><center>The frozenset, from the set Family </center></h1>

• There are two built-in set types, `set` and `frozenset`.<br>
• The set type is `mutable` — the contents can be
changed using methods like `add(..)` and `remove(..)`.<br>
• Since it is `mutable`, it has no `hash` value and cannot
be used as either a `dictionary key` or as an element of another `set`.

```python
x = {1, 2, {11, 22}} # …TypeError: unhashable type: 'set’
        
```

### frozenset
- The frozenset type is **immutable**, and **hashable**:
- its contents *cannot be altered after it is created*;
- it can be used as an element of another set, and as a `dictionary key`.

Like sets, frozensets can be created with the function
`frozenset(..)` with an `iterable` argument (container) :<br>

frozensets can not be created with braces{} <br>

Examples (pay attention to the `un-order` when printed):

```python
x = frozenset('abcd') # arg is a string
print('x =', x) 

y = frozenset([1, 2, 3, 4]) # arg is a list
print('y =', y) 

z = frozenset(set('Iosi')) # arg is a set
print('x =', z)
```

### A frozenset can be member of a set

```python
print(hash(x)) # x the frozy is hashable 

z = { 1, 2, y, x } # x, y are frozensets prev defined
print(z) 
```

### Notes
• To represent `sets of sets`, the `inner sets` must
be `frozensets` objects.<br>
• If `iterable` is not specified, a new `empty set` is
returned. `set()` creates an `empty set`<br>
